```python
import pandas as pd

df = pd.read_csv('data/^GDAXI.csv')
```

```python
df.set_index('Date', inplace = True)
df.dropna(inplace =True)
```

```python
import numpy as np
df['pct_change'] = df.Close.pct_change()
df['log_ret'] = np.log(df.Close) - np.log(df.Close.shift(1))
```

```python

df['price'] = df['pct_change']

df.dropna()
df['price']=(1+df.price).cumprod()
df['logprice']=(df.log_ret).cumsum()
df[['price', 'logprice']].plot()
```

```python
df['price'] = (1+df['price']).cumprod()
df['logprice'] = (1+df['log_ret']).cumsum(axis = 0)
df[['logprice','close']].plot()
```
